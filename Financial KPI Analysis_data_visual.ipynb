```python
import pandas as pd
# Load the Excel file
df = pd.read_excel("startup_financials_dataset.xlsx", engine="openpyxl",sheet_name="Sheet1")
# Display the first few rows
df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Only Date'], df['Revenue (₹)'], label='Revenue (₹)', color='blue', marker='o')
plt.plot(df['Only Date'], df['Operating Expenses (₹)'], label='Expenses (₹)', color='purple', marker='o')
plt.title("Revenue vs Expenses Over Time")
plt.xlabel("Month")
plt.ylabel("₹")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

#This line chart compares monthly revenue and operating expenses. It helps visualize profitability trends
#and identify months where expenses exceeded revenue (negative cash flow).
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Only Date'], df['CAC (₹)'], label='CAC (₹)', color='brown', marker='x')
plt.plot(df['Only Date'], df['LTV (₹)'], label='LTV (₹)', color='blue', marker='x')
plt.title("CAC vs LTV Over Time")
plt.xlabel("Month")
plt.ylabel("₹")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

#This chart tracks Customer Acquisition Cost (CAC) and Customer Lifetime Value (LTV) over time. 
#It helps assess whether your customer acquisition strategy is sustainable and profitable.
```
